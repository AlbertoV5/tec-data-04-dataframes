```python
import pandas as pd
from pathlib import Path
```

```python
# Mapping lets you format an entire DataFrame
file_path = Path.cwd() / "Unsolved" / "Resources" / "sales_data.csv"
sales_df = pd.read_csv(file_path)
sales_df.head()
```

```python
sales_df.dtypes
```

```python
sales_df.columns
```

```python
# Use map() to format:
# Units Sold with comma notation
# Unit Price, Unit Cost, Total Revenue, Total Cost, and Total Profit 
# with a dollar sign, comma notation, and to two decimal places.

sales_df["Units Sold Formatted"] = sales_df["Units Sold"].map("{:,.0f}".format)
sales_df
```
