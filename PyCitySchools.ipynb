```python
# Get average scores of each school by grade
def get_school_grades(complete_df: pd.DataFrame, column: str):
    """Retuns a DataFrame with school names and their 
    {column} score averages by grade.

    Args:
        df (pd.DataFrame): DataFrame with all the data.
        column (str): Column name of the scores.
    
    Returns:
        pd.DataFrame: Grouped and vertically stacked DataFrame.
    """
    df = pd.DataFrame({
        grade: complete_df[complete_df["grade"] == grade
        ].groupby("school_name").mean()[column
        ].map("{:.1f}".format)
        for grade in ("9th", "10th", "11th", "12th")
    })
    df.index.name = ""
    return df

math_scores_df = get_school_grades(complete_df, "math_score")
read_scores_df = get_school_grades(complete_df, "reading_score")

math_scores_df.head()
```

```python
spending_math_scores = per_school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]

spending_reading_scores = per_school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]

spending_passing_math = per_school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]

spending_passing_reading = per_school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]

overall_passing_spending = per_school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
overall_passing_spending
```

```python
f_1 = lambda x: f"{x:.1f}"
f_0 = lambda x: f"{x:.0f}"
spending_summary_df = pd.DataFrame({
    "Average Math Score" : spending_math_scores.map(f_1),
    "Average Reading Score": spending_reading_scores.map(f_1),
    "% Passing Math": spending_passing_math.map(f_0),
    "% Passing Reading": spending_passing_reading.map(f_0),
    "% Overall Passing": overall_passing_spending.map(f_0)})
spending_summary_df
```

```python
size_bins = [0, 999, 1999, 5000]
group_names = ["Small (<1000)", "Medium (1000-1999)", "Large (2000-5000)"]
per_school_summary_df["School Size"] = pd.cut(per_school_counts, size_bins, labels = group_names)
per_school_summary_df.head()
```

```python
def make_summary_df(grouped_df_values: pd.DataFrame) -> pd.DataFrame:
    summary_columns = (
        ("Average Math Score", f_1), 
        ("Average Reading Score", f_1), 
        ("% Passing Math", f_0), 
        ("% Passing Reading", f_0), 
        ("% Overall Passing", f_0)
    )
    return pd.DataFrame(
        {k[0]: grouped_df_values[k[0]].map(k[1])
        for k in summary_columns}
    )

size_summary_df = make_summary_df(per_school_summary_df.groupby(["School Size"]).mean())
size_summary_df
```
