```python
import pandas as pd
from pathlib import Path

resources = Path("resources")
file = resources / "students_complete.csv"
```

```python
df = pd.read_csv(file)
df.head()
```

```python
df_list = df["student_name"].tolist()
df_list
```

```python
for name in df_list:
    if len(name.split()) >= 3:
        print(name.split(), len(name.split()))
```

```python
studends_to_fix = [name for name in df_list if len(name.split()) > 2]
len(studends_to_fix)
```

```python
prefixes = [name.split()[0] for name in studends_to_fix
            if len(name.split()[0]) < 5]
prefixes
```

```python
suffixes = [name.split()[-1] for name in studends_to_fix
            if len(name.split()[-1]) < 4]
suffixes
```

```python
set(suffixes)
```

```python
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
df["student_name"] = df["student_name"].str.replace("|".join(i for i in prefixes_suffixes), "", regex=True)
student_names = df["student_name"].tolist()
```

```python
students_fixed = [name for name in student_names if len(name.split()) >= 3]
len(students_fixed)
```
